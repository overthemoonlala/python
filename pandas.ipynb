```python
import pandas as pd

data = {
    'age': [20,23,48],
    'height': [183,192, 175],
    'weight': [77,83,65]
}

index_name = ['슈퍼맨', '스파이더맨', '베트맨']
data = pd.DataFrame(data, index = index_name)
data
```

```python
#특정열의 값 조회
data['age'][0]
```

# 데이터 프레임 수정

```python
# 열 추가하기

data = pd.DataFrame(data, columns = ['age', 'height', 'weight', 'blood_type'])
data
```

```python
data['blood_type'] = ['A', 'AB', 'B']
data
```

```python
# 행 추가하기
data.loc['헐크'] = [39, 183, 75, 'O']
data
```

```python
# age 열 삭제
data.drop('age', axis =1, inplace = True) # axis =0 (행) , axis =1(열), inplace = True
data
```

```python
# 다시 age 열 추가!
data = pd.DataFrame(data, columns = ['age', 'height', 'weight', 'blood_type'])
```

```python
data['age'] = [10, 20 ,30, 40,]
data
```

```python
data[['age', 'height']]
```

```python
#키가 180보다 큰 사람 조회
data[data['height'] > 180]
```

```python
# 나이가 20살 이상인 사람 조회
data[data['age'] >= 20]
```

```python
# 혈액형이 B인 사람 조회
data[data['blood_type'] == 'A']
```

```python
data.columns
```

```python
#나이가 20살 이상이며 키가 180이상인 사람 조회
data[(data['age'] >= 20) & (data['height'] >= 180)]
```

isin(): 특정 값이 포함된 데이터만을 선택

```python
# 나이가 10살 ,20살!
data[data['age'].isin([10,20])]
```
