# Matplotlib

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [1,2,3]
plt.plot(x,y)
plt.title('My Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi * 10, 500) # p1 + 10 너비에, 500개의 점을 균일하게 찍기
fig, axes = plt.subplots(1,2) #2개의 그래프가 들어가는 Figure 생성
axes[0].plot(x, np.sin(x)) #첫 번째 그래프는 사인(Sin)그래프
axes[1].plot(x, np.cos(x)) #두 번째 그래프는 코사인 그래프
fig.savefig('sin&cos.png')
```
